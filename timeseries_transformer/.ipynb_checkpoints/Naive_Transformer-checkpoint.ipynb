```python
n_customers = 1
batch_size = 32
context_window = 256
forecast_window = 64
head_size = 128
n_heads = 4
device = 'cuda' if torch.cuda.is_available() else 'cpu'

learning_rate = 3e-4
dropout = 0.2

max_iters = 5000
eval_interval = 500
eval_iters = 5

class Head(nn.Module):
    def __init__(self, dimensions, head_size) -> None:
        super().__init__()

        self.key = nn.Linear(dimensions, head_size, bias=False) # dim, head_size
        self.query = nn.Linear(dimensions, head_size, bias=False) # dim, head_size
        self.value = nn.Linear(dimensions, head_size, bias=False) # dim, head_size

        self.dropout = nn.Dropout(dropout)

        self.register_buffer('tril', torch.tril(torch.ones(head_size, head_size))) # not a parameter of the model, so create it with register_buffer.
    
    def forward(self, x):
        b, t, c = x.shape # batch, context+forecast, dim

        k = self.key(x) # batch, context+forecast, head_size
        q = self.query(x) # batch, context+forecast, head_size

        # attention affinities
        weights = q @ k.transpose(-2, -1) * c ** -0.5  # (B, C+F, C+F) # TODO: Rever trecho c**(-0.5)
        weights = weights.masked_fill(self.tril[:t, :t] == 0, float('-inf')) # This is the form of guaranteeing future time tokens do not communicate with former tokens.
        weights = F.softmax(weights, dim=-1)
        
        weights = self.dropout(weights) # batch, context+forecast, context+forecast

        v = self.value(x) # batch, context+forecast, head_size
        out = weights @ v # batch, context+forecast, head_size
        return out

# Multi-head Transformer Model for timeseries Multi-step forecast
class MaskedMultiHeadAttention(nn.Module):
    def __init__(self, head_size, n_heads):
        super().__init__()
        self.heads    = nn.ModuleList([Head(head_size) for _ in range(n_heads)]) # (batch, dim, head_size)
        self.projs    = nn.ModuleList([nn.Linear(head_size, head_size) for _ in range(n_heads)]) # batch, context+forecast, head_size
        self.dropouts = nn.ModuleList([nn.Dropout(dropout) for _ in range(n_heads)])

    def forward(self, x):
        torch.stack(
        out = torch.stack([h(x) for h in self.heads], dim=-1) # Batch, context+forecast, head_size, num_heads
        
        out = torch.stack([p(out[:,:,:,i]) for i,p in enumerate(self.projs)], dim=-1)
        out = torch.stack([d(out[:,:,:,i]) for i,d in enumerate(self.dropouts)], dim=-1)
        
        return out # Batch, context+forecast, head_size, num_heads
```

```python
class NaiveTransformer(nn.Module):
    def __init__(self, head_size = 64, num_heads=5):
        super().__init__()
        
        self.multi_head = MaskedMultiHeadAttentionhead_size, n_heads) # batch, context+forecast, head_size, 
        
        self.lnorm_1 = nn.LayerNorm(head_size) # batch, context+forecast, head_size
        self.lnorm_2 = nn.LayerNorm(head_size) # batch, context+forecast, head_size
        self.lnorm_3 = nn.LayerNorm(head_size) # batch, context+forecast, head_size
        self.lnorm_4 = nn.LayerNorm(head_size) # batch, context+forecast, head_size
        self.lnorm_5 = nn.LayerNorm(head_size) # batch, context+forecast, head_size
        
        self.proj_1 = nn.Linear(head_size, head_size)
        self.proj_2 = nn.Linear(head_size, head_size)
        self.proj_3 = nn.Linear(head_size, head_size)
        self.proj_4 = nn.Linear(head_size, head_size)
        self.proj_5 = nn.Linear(head_size, head_size)
        
        self.quantile_forecasts = nn.ModuleList([nn.Linear(head_size, 3) for i in range(n_heads)])
                                        
    def forward(self, x, target=None):
        x_out = self.multi_head(x)

        x_out_1 = self.lnorm_1(x_out[:,:,:,0])

        x_out_1 = x[:,-5,:,:]  self.proj_1(x_out_1)
        
        x_out_2 = self.lnorm_2(x_out[:,:,:,1])
        x_out_2 = self.proj_2(x_out_2)
        
        x_out_3 = self.lnorm_3(x_out[:,:,:,2])
        x_out_3 = self.proj_3(x_out_3)
        
        x_out_4 = self.lnorm_4(x_out[:,:,:,3])
        x_out_4 = self.proj_4(x_out_4)

        x_out_5 = self.lnorm_5(x_out[:,:,:,4])
        x_out_5 = self.proj_5(x_out_5)

        x_outs = [x_out_1, x_out_2, x_out_3, x_out_4, x_out_5]

        print("Calculating logits")
        logits = torch.stack([qf(_x_out) for qf, _x_out in zip(self.quantile_forecasts, x_outs)], dim=-1) # batch, context+forecast, 3, 5
        
        if target is None:
            loss = None
        else:
            loss = self.quantile_loss_function(logits.squeeze(), target.view(batch_size, -1))

        return logits, loss
        
    def quantile_loss_function(self, predictions, target, quantiles=[0.05, 0.5, 0.95]):
        q = torch.tensor(quantiles).view(-1,1).to(device)
        result = torch.maximum(q*(target - predictions), (1-q)*(predictions - target))
        result = result.sum(dim=-1)
        result = result.mean()
        return result
```
